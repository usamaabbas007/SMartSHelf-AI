# Optimization Prototype

```
from pulp import LpMaximize, LpProblem, LpVariable, lpSum

model = LpProblem(name='shelf-optimization', sense=LpMaximize)
x = LpVariable(name='SKU_BE01', lowBound=0, cat='Integer')
model += x <= 10
model += 276.5 * x
model.solve()
print(f'Status: {model.status}, SKU_BE01 =', x.value())
```
